```python
url = '/content/drive/MyDrive/kaggle/ecommerceDataset.csv'

import pandas as pd

df = pd.read_csv(url)
df.head()
```

```python
new_column_names = ['category', 'description']
df.columns = new_column_names
```

```python
df.head()
```

```python
df.category.value_counts()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.category.replace("Clothing & Accessories", "Clothing_Accessories", inplace=True)
df.category.unique()
```

```python
df['category'] = "__label__" + df['category'].astype(str)
```

```python
df.head()
```

```python
df['category_description'] = df['category'] + " " + df['description']
df.head()
```

```python
df['category_description'][0]
```

```python
import re

def preprocess(text):
    text = re.sub(r'[^\w\s\']',' ', text)
    text = re.sub(' +', ' ', text)
    return text.strip().lower()
```

```python
df['category_description'] = df['category_description'].map(preprocess)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
```

```python
train.head(3)
```

```python
train.shape, test.shape
```

```python
train.to_csv("ecommerce.train", columns=["category_description"], index=False, header=False)
test.to_csv("ecommerce.test", columns=["category_description"], index=False, header=False)
```

```python
import fasttext

model = fasttext.train_supervised(input="ecommerce.train")
model.test("ecommerce.test")
```

```python
model.predict("Enhanced Graphics Rendering: This application harnesses the power of dedicated GPU processing to elevate visual experiences, delivering stunning graphics, smoother frame rates, and seamless rendering in video games, multimedia editing, and complex visual simulations.")
```

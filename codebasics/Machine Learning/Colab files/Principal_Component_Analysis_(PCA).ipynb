```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
X = df
y = dataset.target
```

```python
X.head()
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA
# pca is dimensionality reduction it may lower your accuracy
pca = PCA(0.95) # give me the 95 percent useful information

X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

0.14 (14%) percent getting useful of information, 0.136 (13.6%) getting useful of info, so on...*italicized text*

```python
 pca.n_components_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df[df.MaxHR > (df.MaxHR.mean()+3*df.MaxHR.std())]
```

```python
df[df.FastingBS > (df.FastingBS.mean()+3*df.FastingBS.std())]
```

```python
df[df.Oldpeak>(df.Oldpeak.mean()+3*df.Oldpeak.std())]
```

```python
df2 = df1[df1.Oldpeak <= (df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
```

```python
df2.shape
```

```python
df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())]
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
df.ChestPainType.unique()
```

```python
df.RestingECG.unique()
```

```python
df.ExerciseAngina.unique()
```

```python
df.ST_Slope.unique()
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down':1,
        'Flat':2,
        'Up':3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal':1,
        'ST':2,
        'LVH': 3
    },
    inplace=True
)

df4.Sex.replace(
    {
        'M':1,
        'F':0
    },
    inplace=True
)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True)
df5.head()
```

```python
X = df5.drop('HeartDisease', axis=1)
y = df5.HeartDisease

X.head()
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
model_rf.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

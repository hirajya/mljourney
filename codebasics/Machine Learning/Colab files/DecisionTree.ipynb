```python
import pandas as pd
```

```python
# Read the CSV file
csv_file_path = '/content/drive/MyDrive/datasets/salaries.csv'
df = pd.read_csv(csv_file_path)

# Perform operations on the dataframe
df.head()
```

```python
inputs = df.drop('salary_more_then_100k', axis=1)
inputs[0:5]
```

```python
target = df['salary_more_then_100k']
target[0:5]
```

```python
from sklearn.preprocessing import LabelEncoder

le_company = LabelEncoder()
le_job= LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company', 'job', 'degree'], axis=1)
inputs_n.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2, random_state=42)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[2, 2, 1]])
```

```python
# Read the CSV file
csv_file_path = '/content/drive/MyDrive/datasets/titanic.csv'
df = pd.read_csv(csv_file_path)

# Perform operations on the dataframe
df.head()
```

```python
# drop unneccessary columns
df_titanic = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1)
```

```python
# make all column names lower case
df_titanic.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
# check null values
null_counts = df_titanic.isnull().sum()
null_counts
```

```python
# apply median for age null values

age_median = df_titanic['age'].median()
df_titanic['age'].fillna(age_median, inplace=True)
```

```python
# checking null values after cleaning
df_titanic.isnull().sum()
```

```python
# change the sex column to numerical categorical values
from sklearn.preprocessing import LabelEncoder

le_age = LabelEncoder()
df_titanic['sex'] = le_age.fit_transform(df_titanic['sex'])
```

```python
df_titanic.head()
```

```python
df_titanic.shape
```

```python
X = df_titanic.drop(['survived'], axis=1)
y = df_titanic.survived
```

```python
X[0:5]
```

```python
y[0:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
